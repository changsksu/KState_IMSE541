## Load The Dataset or File From My Google Drive by export link

```python
def get_export_link(share_url, file_format="csv"):
    """
    Convert a Google Sheets share URL to an export link.

    share_url: str, Google Sheets share link
    file_format: "csv" or "xlsx"
    """
    # Extract the file ID (between '/d/' and '/edit')
    try:
        file_id = share_url.split("/d/")[1].split("/")[0]
    except IndexError:
        raise ValueError("Invalid Google Sheets link")

    # Build export link
    export_link = f"https://docs.google.com/spreadsheets/d/{file_id}/export?format={file_format}"
    return export_link


# Example usage
share_url = "https://docs.google.com/spreadsheets/d/1x3ITzHSbhm07Fl3CqfKadVjjQmVFpDln/edit?usp=drive_link&ouid=106031054461004059338&rtpof=true&sd=true"

excel_link = get_export_link(share_url, "xlsx")

print("Excel export link:", excel_link)
```

```python
import pandas as pd

# Read CSV file
df = pd.read_csv("/content/iris_small.csv")

# Select one column: sepal length (cm)
sepal_length = df["sepal length (cm)"]

# Mean
# the function .mean() is used to calculate the average value of a column
# (or of each column if applied to a whole DataFrame).

mean_val = sepal_length.mean()   # average value
print("Mean (sepal length):", mean_val)

# Variance
var_val = sepal_length.var()     # measure of spread (squared units)
print("Variance (sepal length):", var_val)

# Standard Deviation
std_val = sepal_length.std()     # square root of variance
print("Standard Deviation (sepal length):", std_val)

# Percentiles (Quartiles)
q25 = sepal_length.quantile(0.25)   # quantile(0.25)  25% of the data are below this value (Q1)
q50 = sepal_length.quantile(0.50)   # quantile(0.50)  same as the Median (Q2)
q75 = sepal_length.quantile(0.75)   # quantile(0.75)  75% of the data are below this value (Q3)

print("25th percentile (Q1):", q25)
print("50th percentile (Median):", q50)
print("75th percentile (Q3):", q75)
```

If you want all results shown with 3 decimal places, you can use Pythons round() function or formatted printing.

```python
# Mean
mean_val = round(sepal_length.mean(), 3)   # average value
print("Mean (sepal length):", mean_val)

# Variance
var_val = round(sepal_length.var(), 3)     # measure of spread (squared units)
print("Variance (sepal length):", var_val)

# Standard Deviation
std_val = round(sepal_length.std(), 3)     # square root of variance
print("Standard Deviation (sepal length):", std_val)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read CSV file
df = pd.read_csv("/content/iris_small.csv")

# Select one column: sepal length (cm)
sepal_length = df["sepal length (cm)"]

# Calculate mean
mean_val = sepal_length.mean()

# Histogram with more bins
plt.hist(sepal_length, bins=10, color="lightgreen", edgecolor="black")
plt.title("Histogram of Sepal Length (cm) with More Bins")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")

# Add vertical line for mean
plt.axvline(mean_val, color='red', linestyle='dashed', linewidth=2, label=f"Mean = {mean_val:.2f}")
plt.legend()

plt.show()
```

```python
# Read CSV file
df = pd.read_csv("/content/iris_small.csv")

# Select one column: sepal length (cm)
sepal_length = df["sepal length (cm)"]

# Calculate mean
# Function to create a stem-and-leaf plot
def stem_and_leaf(data):
    # Dictionary to store stems and their leaves
    stem_dict = {}

    # Process each value
    for value in sorted(data):
        # Split into stem (integer) and leaf (decimal part, rounded to one digit)
        stem = int(value)
        leaf = int((value - stem) * 10)  # Get first decimal digit
        if stem not in stem_dict:
            stem_dict[stem] = []
        stem_dict[stem].append(leaf)

    # Print the stem-and-leaf plot
    print("Stem | Leaf")
    print("-----|-----")
    for stem in sorted(stem_dict.keys()):
        leaves = " ".join(str(leaf) for leaf in sorted(stem_dict[stem]))
        print(f"{stem}    | {leaves}")

# Generate the stem-and-leaf plot
stem_and_leaf(sepal_length)
```
